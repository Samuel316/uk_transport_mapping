```python
# Importing Libraries

import folium
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

from shapely.geometry import Point
from convertbng.util import convert_bng
```

```python
df_stations = pd.read_csv("Data/GB stations.csv")

df_stations["Eastings"], df_stations["Northings"] = convert_bng(df_stations.Longitude, df_stations.Latitude)
gdf_stations = gpd.GeoDataFrame(df_stations, geometry=gpd.points_from_xy(df_stations.Longitude, df_stations.Latitude))

gdf_stations.head(3)
```

```python
# OpenStreetMap
map = folium.Map(location = [52, 0], tiles='OpenStreetMap' , zoom_start = 8)

for coordinates, name, owner in zip(zip(gdf_stations.geometry.y, gdf_stations.geometry.x), gdf_stations.Station, gdf_stations.Owner):
    map.add_child(folium.Marker(location = coordinates, popup=f"{name}:<br>{owner}"))
                                
map
```

```python
map.save("example.html")
```
