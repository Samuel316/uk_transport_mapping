```python
from pathlib import Path
import matplotlib.pyplot as plt
import geopandas
```

```python
path = Path("Data/Border_railway.geojson")
assert path.is_file()
```

```python
gdf = geopandas.read_file(path)
```

```python
gdf.columns
```

```python
fig, ax = plt.subplots(figsize=(20,20))

gdf.loc[[0], "geometry"].plot(ax=ax, label=len(gdf.loc[[0], "geometry"][0].coords))
gdf.loc[[0], "geometry"].simplify(0.00001).plot(ax=ax, color="C1", label=len(gdf.simplify(0.00001)[0].coords))
gdf.loc[[0], "geometry"].simplify(0.0001).plot(ax=ax, color="C2", label=len(gdf.simplify(0.0001)[0].coords))
gdf.loc[[0], "geometry"].simplify(0.001).plot(ax=ax, color="C3", label=len(gdf.simplify(0.001)[0].coords))
gdf.loc[[0], "geometry"].simplify(0.01).plot(ax=ax, color="C4", label=len(gdf.simplify(0.01)[0].coords))

plt.legend()
```
